# Date time Series

```python
import pandas as pd
```

```python
df = pd.read_csv('ETH_1H.csv')
```

```python
#check if date is in date time fomat
df.loc[0,'Date'].day_name()
```

```python
#convert to DATETIME FOMAT
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# If It throws an error You Have To Specify Datetime fomat
```

```python
df['Date'] = pd.to_datetime(df['Date'], format = '%Y-%m-%d %I-%p')
# This can also be done when opening file at the top -- > As Shown BELOW

# d_parser = lambda x: pd.datetime.strptime(x,'%Y-%m-%d %I-%p')
# df = pd.read_csv('ETH_1H.csv', parse_dates=['Date'], date_parser = d_parser)
```

```python
# Check Day Name of a certern row value
df.loc[0,'Date'].day_name()
```

```python
# Day Names for a whole series
df['Date'].dt.day_name()
```

```python
# Add A new Column of all week days for the  dates
df['New Day'] = df['Date'].dt.day_name()
df
```

```python
# Ealiers Date 

df['Date'].min()
```

```python
# Most Recent Date
df['Date'].max()
```

```python
#Subract Dates
df['Date'].max() - df['Date'].min()
```

```python
# Filters By DAte
```

```python
# Filter Trades for Only 2019 --- Using Strings
d_filter = (df['Date'] >= '2019') & (df['Date'] < '2020')
df.loc[d_filter]
```

```python
# Filter Trades for Only 2019 --- Using to_datetime
d2_filter = (df['Date'] >= pd.to_datetime('2019-01-01')) & (df['Date'] < pd.to_datetime('2020-01-01'))
df.loc[d2_filter]
```
